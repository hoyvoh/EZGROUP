```python
# Get news
response = requests.get('https://cafef.vn/bat-dong-san.chn')
response.text
```

```python
bs = BeautifulSoup(response.text, 'html.parser')
links = [a['href'] for div in bs.find_all(class_="tlitem box-category-item") for a in div.find_all("a")]
links = list(set(links))
links
```

```python
pages = []
for link in links:
    pages.append(requests.get(str('https://cafef.vn/'+link)).text)
pages[0]
```

```python
# news: title, cate, created_at, content, img_list
news = []
import regex as re

for page in pages:
    bs = BeautifulSoup(page, 'html.parser')
    page_content = bs.find('div', {'class': 'left_cate totalcontentdetail'})

    # Parse the specific section again if needed
    bs = BeautifulSoup(str(page_content), 'html.parser')

    # Extracting elements
    title = bs.find('h1', {'class': 'title'}).get_text().strip() if bs.find('h1', {'class': 'title'}) else ''
    date = bs.find('span', {'class': 'pdate'}).get_text().strip() if bs.find('span', {'class': 'pdate'}) else ''
    cate = bs.find('a', {'class': 'cat'}).get_text().strip() if bs.find('a', {'class': 'cat'}) else ''
    
    # Separating paragraphs by [SEP] in the content
    content_div = bs.find('div', {'class': 'w640 fr clear'})
    content = '[SEP]'.join(paragraph.strip() for paragraph in content_div.get_text().split('\n') if paragraph.strip()) if content_div else ''
    
    # Joining image sources and labels
    imgs = '[SEP]'.join([img['src'] for img in bs.find_all('img') if 'src' in img.attrs])
    labs = '[SEP]'.join([img['title'] for img in bs.find_all('img') if 'title' in img.attrs])
    block = BeautifulSoup(str(bs.find('div', {'class':'t-contentdetail content_source'})), 'html.parser')
    author = block.find('p', {'class':'author'}).get_text().strip()
    source = block.find('p', {'class':'source'}).get_text().strip()
    
    
    # Constructing the dictionary
    new = {
        'title': title,
        'date': date,
        'category': cate,
        'content': content,
        'images': imgs,
        'labels': labs,
        'author':author,
        'source':source
    }
    
    news.append(new)

pd.DataFrame(news).to_csv('news.csv', index=False, encoding='utf8')
```

```python
# Blogs
url = 'https://blogbatdongsan.vn/'
response = requests.get(url)
response.text
```

```python
bs = BeautifulSoup(response.text, 'html.parser')
content_area = bs.find('div', class_='content-area')
links = [a['href'] for a in content_area.find_all('a', href=True)]

print(links)
```

```python
# title, cate, 
blogs =[]

for link in links:
    try:
        page = requests.get(link)
        page.raise_for_status()  # To raise an exception for bad responses (e.g. 404, 500)

        bs = BeautifulSoup(page.text, 'html.parser')

        # Extract the author, if it exists
        author = bs.find('h5', {'class': 'author-name uppercase pt-half'})
        author_name = author.get_text().strip() if author else 'Unknown Author'

        # Extract the post section
        post = bs.find('div', {'class': 'large-8 col'})
        if post is None:
            continue  # Skip this link if no post section is found
        
        # Convert post section to a new BeautifulSoup object for further parsing
        bs_post = BeautifulSoup(str(post), 'html.parser')

        # Extract category, title, date, and content
        cate = bs_post.find('h6').get_text().strip() if bs_post.find('h6') else 'No Category'
        title = bs_post.find('h1').get_text().strip() if bs_post.find('h1') else 'No Title'
        date = bs_post.find('span', {'class': 'post-meta is-small op-8'})
        date_text = date.get_text().strip() if date else 'No Date'
        
        # Extract content, split paragraphs, and join them with [SEP]
        content = '[SEP]'.join([p.strip() for p in bs_post.find('div', {'class': 'entry-content single-page'}).get_text().strip().split('\n') if p.strip()])
        
        # Extract image links and alt texts
        imgs = '[SEP]'.join([img['data-src'] for img in bs_post.find('div', {'class': 'entry-content single-page'}).find_all('img') if 'data-src' in img.attrs])
        alts = '[SEP]'.join([img['alt'] for img in bs_post.find('div', {'class': 'entry-content single-page'}).find_all('img') if 'alt' in img.attrs])

        # Create the blog dictionary to store the scraped data
        blog = {
            'author': author_name,
            'category': cate,
            'title': title,
            'date': date_text,
            'content': content,
            'images': imgs,
            'alt_texts': alts
        }

        # Append the blog data to the blogs list
        blogs.append(blog)

    except requests.exceptions.RequestException as e:
        print(f"Request error for {link}: {e}")
    except Exception as e:
        print(f"An error occurred while processing {link}: {e}")
    
pd.DataFrame(blogs).to_csv('blogs.csv', index=False, encoding='utf8')
```
